# Grouping dataset into genres

```python
import pandas as pd
import pathlib
import os
```

## Tracks metadata

```python
df_tracks = pd.read_csv('../../datasets/fma/fma_metadata/tracks.csv', header=1)
df_tracks = df_tracks.rename({"Unnamed: 0": 'track_id'}, axis='columns')
df_tracks = df_tracks.drop(labels=[0], axis=0)
```

```python
df_tracks.columns
```

```python
df_tracks = df_tracks[["track_id", "id", "title", "tracks", 'type', "associated_labels", "name", "split", "subset", "composer", "genre_top", "genres", "genres_all"]]
df_tracks["track_id"] = df_tracks["track_id"].astype(str)
df_tracks.index = df_tracks["track_id"]
df_tracks = df_tracks.drop(columns="track_id")
df_tracks.head(20)
```

## Genres metadata 

```python
df_genres = pd.read_csv('../../datasets/fma/fma_metadata/genres.csv', index_col='genre_id')
df_genres.head(20)
```

```python
df_genres.loc[[2]]["title"].values[0]
```

## Create genres folders

```python
genres_count = {}

rootdir = '../../datasets/fma/fma_small/'
proj_path = os.path.abspath(os.getcwd())[:-11]
dataset_path = os.path.join(proj_path, 'datasets', 'fma', 'fma_small')

for subdir, dirs, files in os.walk(rootdir):
    for file in files:
        if file.endswith('.mp3'):

            track_id = file[:-4].lstrip('0')
            genre = df_tracks.loc[track_id]["genre_top"]
            genres_count[genre] = genres_count.get(genre, 0) + 1

            old_path = os.path.join(dataset_path, subdir[-3:], file)
            new_path = os.path.join(dataset_path, genre, file)

            pathlib.Path(os.path.join(dataset_path, genre)).mkdir(parents=True, exist_ok=True)
            os.rename(old_path, new_path)
```
